```python
# Импорт необходимых библиотек.
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Вызов файла.
path_to_db = '/Users/yanako/MyDrive/COURSES/YANDEX_ANALITICS/datasets_ya/telecomm_csi.db' 
engine = create_engine(f'sqlite:///{path_to_db}', echo = False)
```

```python
# Составление запросадля базы данных.
query = '''SELECT u.user_id,
                  u.lt_day,
                  CASE
                      WHEN u.lt_day <= 365 THEN 'new'
                      ELSE 'not new'
                  END AS is_new,
                  u.age,
                  CASE
                      WHEN u.gender_segment > 0.5 THEN 'female'
                      WHEN u.gender_segment < 0.5 THEN 'male'
                      ELSE NULL
                  END AS gender,
                  u.os_name,
                  u.cpe_type_name,
                  lo.country,
                  lo.city,
                  a.title AS age_segment,
                  t.title AS traffic_segment,
                  li.title AS lifetime_segment,
                  u.nps_score,
                  CASE
                      WHEN u.nps_score < 7 THEN 'detractors'
                      WHEN u.nps_score > 8 THEN 'promoters'
                      ELSE 'passives'
                  END AS nps_group
           FROM user AS u
           LEFT JOIN location AS lo ON u.location_id=lo.location_id
           LEFT JOIN age_segment AS a ON u.age_gr_id=a.age_gr_id
           LEFT JOIN traffic_segment AS t ON u.tr_gr_id=t.tr_gr_id
           LEFT JOIN lifetime_segment AS li ON u.lt_gr_id=li.lt_gr_id
        '''
```

```python
# Создание датафрейма.
df = pd.read_sql(query, engine)
```
